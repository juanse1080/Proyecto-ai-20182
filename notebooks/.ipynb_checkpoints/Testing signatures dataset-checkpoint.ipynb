```python
from PIL import Image
#from scipy.signal import convolve2d
from sklearn.model_selection import train_test_split
import imageio as imgeio
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score, KFold
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.metrics import *
```

```python
class Data:
    def __init__(self, reales, falsas, grilla):
        self.data = [imgeio.imread(reales), imgeio.imread(falsas)]
        self.grid = grilla
        self.n = min(imgeio.imread(reales).shape[1], imgeio.imread(falsas).shape[1])/grilla[0]
        self.m = min(imgeio.imread(reales).shape[0], imgeio.imread(falsas).shape[0])/grilla[1]
        
    def pintar(self, size): #size es una tupla
        fig = plt.figure(figsize=size)
        for item,i in enumerate(self.data):
            fig.add_subplot(1, 2, item+1)
            plt.imshow(i, cmap=plt.cm.Greys_r)
            plt.title(item+1)
            plt.axis("off")
            
    def dibujarFirmas(self, x, y):
        fig = plt.figure(figsize=(15,20)) 
        for i,img in enumerate(x):
            fig.add_subplot(8, 4, i+1)
            plt.imshow(img.reshape(self.shape), cmap=plt.cm.Greys_r)
            plt.title('Clase: '+str(y[i]))
            plt.axis("off")
    
    def cargarFirmas(self,borde):
        cont = 0
        x = np.ones((32,(self.n-2*borde)*(self.m-2*borde)))
        for k in self.data:
            datos = np.mean(k,axis=2)
            for i in range(self.grid[0]):
                for j in range(self.grid[1]):
                    temp = datos[self.m*j+borde:self.m*(j+1)-borde,self.n*(i)+borde:self.n*(i+1)-borde]
                    x[cont] = temp.reshape(1,(self.n-2*borde)*(self.m-2*borde))
                    cont = cont +1
        self.shape = temp.shape
        y = np.array([1 if i < 8 else 0 for i in range(32)])
        return x, y
        
    def filtro(self, f, (x, y)):
        fig = plt.figure(figsize=(15,20)) 
        data = np.ones((32,(self.shape[0]-2)*(self.shape[1]-2)))
        for i,img in enumerate(x):
#             fig.add_subplot(8, 4, i+1)
            filtro = convolve2d(img.reshape(self.shape), f, mode="valid")
            data[i] = filtro.reshape(1, (self.shape[0]-2)*(self.shape[1]-2))
#             plt.imshow(filtro, cmap=plt.cm.Greys_r)
#             plt.title('Clase: '+str(y[i]))
#             plt.axis("off")
        return data, y
        
    def PCA(self, clasificador, (x, y)):
        n_components = range(10,200,10)
        means, stds = [], []
        for n in n_components:
            p = Pipeline((("pca", PCA(n_components=n)), ("classifier", clasificador) ))
            s = cross_val_score(p, x, y, cv=KFold(5, shuffle=True), scoring=make_scorer(accuracy_score))
            means.append(np.mean(s))
            stds.append(np.std(s))
        means = np.r_[means]
        stds  = np.r_[stds]
        return n_components, means, stds
#         plt.plot(n_components, means, color="black")
#         plt.fill_between(n_components, means-stds, means+stds, color="blue", alpha=.5)
#         plt.xlabel("number of PCA componets")
#         plt.ylabel("accuracy")
            
    def entrenar(self, clasificador):
        self.model = clasificador
        self.model.fit(self.x_train, self.y_train)
        print "acierto en train", self.model.score(self.x_train, self.y_train)
        
    def predecir(self):
        self.model.predict(self.x_test)
        print "acierto en test", self.model.score(self.x_test, self.y_test)
    
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
firmas = Data('../data/real.jpeg','../data/falsa.jpeg', (4,4))
data = firmas.cargarFirmas(20)
f = [[0, 1, 0],
     [1, 4, 1],
     [0, 1, 0]]
cont = 0
fig = plt.figure(figsize=(15,3)) 
clasificators = {'Support Vector Machine':SVC(),'Gaussian Naive Bayes':GaussianNB(),'Decision Tree classifier':DecisionTreeClassifier()}
for key, classificador in clasificators.items():
    cont = cont +1
    n_components,means,stds = firmas.PCA(classificador, data)
    fig.add_subplot(1, len(clasificators), cont)
    plt.plot(n_components, means, color="black")
    plt.fill_between(n_components, means-stds, means+stds, color="blue", alpha=.5)
    plt.title(key)
    plt.xlabel("number of PCA componets")
    plt.ylabel("accuracy")
#firmas.PCA(SVC(), firmas.filtro(f, data))
```
